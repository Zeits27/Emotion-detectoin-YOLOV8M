```python
# =========================
# 03_evaluation.ipynb
# =========================

import os
import matplotlib.pyplot as plt
from ultralytics import YOLO
import seaborn as sns

# 1. Load model hasil training
model_path = "../weight/best5.pt"   # ganti sesuai lokasi
model = YOLO(model_path)

# 2. Evaluasi di validation/test set
results = model.val(
    data="../data2/augmented/data.yaml",   # dataset config
    split="test",        # bisa 'val' atau 'test'

)
```

```python

# 3. Print metrik utama
print("Precision:", results.results_dict['metrics/precision(B)'])
print("Recall:", results.results_dict['metrics/recall(B)'])
print("mAP50:", results.results_dict['metrics/mAP50(B)'])
print("mAP50-95:", results.results_dict['metrics/mAP50-95(B)'])
```
